## Task1  Merging all the results 

```python
df = pd.read_csv(r'/Users/alix/Documents/climbing_data/combined_results.csv')

files = [file for file in os.listdir(r'/Users/alix/Documents/climbing_data')]

all_results = pd.DataFrame()
for file in files:
    df = pd.read_csv(r'/Users/alix/Documents/climbing_data/'+ file)
    all_results = pd.concat([all_results, df])
all_results.to_csv("all_data.csv", index= False)
```
